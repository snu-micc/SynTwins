```python
from rdkit import Chem
from syntwins.syntree import SynTree
from syntwins.data_utils import SynTwin_DataLoader

loader = SynTwin_DataLoader(data_dir='data/', _nBits=1024)
```

```python
def run_syntwins(target_smi, _nBits=1024, d=3, k=10):
    print ('Target molecule:', target_smi)
    display(Chem.MolFromSmiles(target_smi))
    tree = SynTree(loader, _nBits=_nBits, k=k, max_depth=d, verbose=True)
    analogs = tree.get_twins(target_smi)
    tree.show_analogs()
    return

target_smi = 'OC(C1=CC=CC(C2=NC(C=CC=C3)=C3N2CC4=CC=CC=C4)=C1)=O'
run_syntwins(target_smi)
```
